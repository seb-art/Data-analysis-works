# 1.Hero Technologies Inc. (HENC) data extraction

```python
from yahoo_fin.stock_info import get_data
import matplotlib.pyplot as plt

#here I have created function for extracting data of different companies using yahoo API
#This function takes the argument company name
def Monthly_data(company_name):
     monthly_data = get_data(company_name, start_date="09/12/2021",
                         end_date="09/12/2023", index_as_date = True,
                         interval="1mo")
     return monthly_data   #Here we are extracting the data

#calling the function Monthly data to extract the data of company Hero Technologies usign tiket HENC
print(Monthly_data("HENC"))
```

```python
#created the function for checking high monthly collection of each company through two years
def High_monthly_collection(company_name):
        Monthly_data(company_name)['high'].plot(title="High Monthly Collection")
        plt.show() #It shows the high level of stock

#created the function for checking Low monthly collection of each company through two years
def Low_monthly_collection(company_name):
        Monthly_data(company_name)['low'].plot(title="Low Monthly Collection")
        plt.show() #it shows the low level of stock

#calling the function for checking high monthly collection data plot
High_monthly_collection("HENC")

#calling the function for checking low monthly collection data plot
Low_monthly_collection("HENC")
```

```python
import pandas as pd
```

```python

#created method to check the graphs for two columns together
#This method takes the inputs as campany name, column 1 name. column 2 name, column 1 color , column 2 color, description of the graph

def relationship_between_diff_columns(company_name,column1,column2,column1_color,column2_color,description):
      df = Monthly_data(company_name)
      fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 8))
      axes[0].plot(df.index, df[column1], label=column1, color=column1_color)
      axes[0].plot(df.index, df[column2], label=column2, color=column2_color)
      axes[0].set_ylabel(description)
      axes[0].set_title(description)
      axes[0].legend()
      axes[1].set_xlabel('Date')
      plt.tight_layout()
      plt.show()

#Created method to check the correlation between two columns
def correlation_btw_columns(company_name,column1,column2,description):
     df = Monthly_data(company_name)
     correlation = df[column1].corr(df[column2])
     print(description,correlation)

#calling the function to check the graph between volume and close
relationship_between_diff_columns("HENC",'volume','close','orange','green','Volume vs Close')

#Calling the functions to check the correlation between volume and close column
correlation_btw_columns("HENC",'volume','close','Correlation between volume and close graph is: ')
```

```python
#Callig the function to check the graph between  Open and CLose column
relationship_between_diff_columns("HENC",'open','close','blue','brown','Open VS Close')

#Calling the function to check the correlation between Open and close column
correlation_btw_columns("HENC",'open','close','Correlation between Open and Close Graph')
```

```python
#Creating the function for checking least and most transcation
def Show_most_and_least_transaction(company_name,color):
      df = Monthly_data(company_name)
      df['open'].plot(kind='bar', color = color)
      plt.title('Monthly Transaction Count')
      plt.xlabel('Month')
      plt.ylabel('Transaction Count')
      plt.show()
      mostTransactionMonth = df['open'].idxmax()#Most transaction month
      leastTransactionMonth = df['open'].idxmin()
      print("Most Transaction month is :",mostTransactionMonth)
      print("Least transaction month is :",leastTransactionMonth)

#Calling the function to check the monthly transactions using bar graph
Show_most_and_least_transaction("HENC","orange")

```
